```python
#Kalbert
# 
# 

DURUM = " "

class Nokta: 
    
    def __init__(self, x, y):
        self.x=x
        self.y=y
             
    def mesafe(self,nokta):
        import math
        return math.sqrt((math.pow((self.x-nokta.x),2) + math.pow((self.y-nokta.y),2))* 1.0)
    
        
class Dogru:  
    
    def __init__(self,nokta1, nokta2):
        self.nokta1 = Nokta(nokta1.x, nokta1.y)
        self.nokta2 = Nokta(nokta2.x, nokta2.y)

    def dogru_kesisim(self, dogru2):
        X1, Y1 = self.nokta1.x, self.nokta1.y                      
        X2, Y2 = self.nokta2.x, self.nokta2.y
        X3, Y3 = dogru2.nokta1.x, dogru2.nokta1.y
        X4, Y4 = dogru2.nokta2.x, dogru2.nokta2.y
        
        ua_pay=((X4-X3)*(Y1-Y3)-(Y4-Y3)*(X1-X3))                      
        ub_pay=((X2-X1)*(Y1-Y3)-(Y2-Y1)*(X1-X3))
        uab_pay=((Y4-Y3)*(X2-X1)-(X4-X3)*(Y2-Y1))      
        if (uab_pay==0) and (ua_pay==0) and (ub_pay==0): 
            global DURUM
            DURUM ="Dogrular çakışıyor, Çember oluşturulmaz"    
            return False
        else:        
            return True
        
class Cember:
    
    def __init__(self, nokta1, nokta2,nokta3):
        self.nokta1 = Nokta(nokta1.x, nokta1.y)
        self.nokta2 = Nokta(nokta2.x, nokta2.y)
        self.nokta3 = Nokta(nokta3.x, nokta3.y)      
    def merkez_yaricap(self):
        x1, y1 = self.nokta1.x, self.nokta1.y  
        x2, y2 = self.nokta2.x, self.nokta2.y
        x3, y3 = self.nokta3.x, self.nokta3.y  
       
        D1=Dogru(Nokta(x1,y1),Nokta(x2,y2))  
        D2=Dogru(Nokta(x2,y2),Nokta(x3,y3))
        
        DogrularKesisimDurumu=D1.dogru_kesisim(D2)
       
        if DogrularKesisimDurumu==False:
            return False            
        maBolunen= y2-y1
        maBolucu= x2-x1
        mbBolunen= y3-y2
        mbBolucu= x3-x2            
       
        ma=maBolunen/maBolucu
        mb=mbBolunen/mbBolucu
        
        global DURUM
        if mb-ma==0.0:
            DURUM="Doğrular paraleldir, daire oluşmaz"
            return False
        
        x=(ma*mb*(y1-y3)+mb*(x1+x2)-ma*(x2+x3))/(2*(mb-ma))                  
        Ya=(-1/ma)*(x-((x1+x2)/2))+((y1+y2)/2)
        Yb=(-1/mb)*(x-((x2+x3)/2))+((y2+y3)/2)
        
        DURUM="Çemberin Merkezi = ( {},{})\n".format(x,Ya)  
        yaricap=self.nokta1.mesafe(Nokta(x,Yb))          
        DURUM=DURUM + "\nÇemberin Yarıçapını = {}".format(yaricap)
        
P1=Nokta(6,3)  
P2=Nokta(4,1)          
P3=Nokta(2.59,1.59)        

Cember(P1,P2,P3).merkez_yaricap()
print(DURUM)
 
```
